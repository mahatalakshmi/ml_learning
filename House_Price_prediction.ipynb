```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
```

```python
new_data=df.drop(['date','street','statezip'],axis=1)
```

```python
b=(new_data.columns)
```

```python
new_data['city']=pd.factorize(new_data['city'])[0]
new_data['country']=pd.factorize(new_data['city'])[0]
```

```python
from sklearn.preprocessing import StandardScaler
a=StandardScaler()
new_data=a.fit_transform(new_data)
new_data=pd.DataFrame(new_data,columns=b)
```

```python
x=new_data.drop('price',axis=1)
y=new_data[['price']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred= lr.predict(x_test)
y_pred.shape
y_test.shape
```

```python
print("Intercept: ", lr.intercept_)
print("Coefficient: ", lr.coef_)
```

```python
lr.score(x_test,y_test)
```
